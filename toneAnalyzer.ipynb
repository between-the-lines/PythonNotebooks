#### Extracting valid tweets from the twitter feed

```python
validTweets = []  # The list of all valid tweets
for i in range(len(allTweets)):
    try:
        tweet = json.loads(allTweets[i])  # tweet is a dictionary
        text = tweet["text"]
        textLower = text.lower()
        if (" s8" in textLower) and \
        ("giveaway" not in textLower) and \
        ("giving away" not in textLower) and \
        (" win " not in textLower) and \
        ("look like a supersized version" not in textLower) and \
        ("I can't wait to try Samsung Dex" not in textLower) and \
        ("bestbuy" not in textLower) and \
        ("technews" not in textLower) and \
        (("samsung" in textLower) or "galaxy" in textLower):
            print(text)
            validTweets.append(text)
    except:
        pass
print(len(validTweets))
```

#### Processing valid tweets in Watson Tone Analyzer

```python
import json
from watson_developer_cloud import ToneAnalyzerV3

tone_analyzer = ToneAnalyzerV3(
    username='YOUR SERVICE USERNAME',
    password='YOUR SERVICE PASSWORD',
    version='2016-02-11')

# print(json.dumps(tone_analyzer.tone(text='Samsung Galaxy S8 International Giveaway!'), indent=2))
tweetWatsonTupList = []
for i in range(len(validTweets)):
# for i in range(300,310):
    try:
        watsonStr = json.dumps(tone_analyzer.tone(text=validTweets[i]))
        tweetWatsonTupList.append((validTweets[i], watsonStr))
    except:
        pass
```

```python
g = open("jsonList.json", 'w')
for i in range(len(tweetWatsonTupList)):
    g.write((tweetWatsonTupList[i])[1])
    g.write("\n")
g.close()
```
